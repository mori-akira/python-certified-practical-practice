```python
# 文字列のチェックメソッド
print('hoge'.isalnum())
print('hoge+fuga'.isalnum())
print()

print('hoge'.isalpha())
print('h0ge'.isalpha())
print('あいうえお'.isalpha())
print()

tmp = lambda s: print(s.ljust(10), s.isdecimal(), s.isdigit(), s.isnumeric())
tmp('123')
tmp('-123')
tmp('123.456')
tmp('１２３')
tmp('①②③')
tmp('ⅠⅡⅢ')
tmp('百二十三')
print()

print('aaa'.isidentifier())
print('123'.isidentifier())
print('___'.isidentifier())
print('_%_'.isidentifier())
print()

print('abcde'.islower())
print('ABCDE'.isupper())
print('ABcde'.islower())
print()

print('   '.isspace())
print('\t\t\t'.isspace())
print('\r\n'.isspace())
print('  xx  xx '.isspace())
print()

print('Title'.istitle())
print('TITLE'.istitle())
print('title.'.istitle())
```

```python
# 文字の変換
print('HoGe'.upper())
print('HoGe'.lower())
print('HoGe'.swapcase())
print('HoGe'.capitalize())
print('AbCd Efgh IJKL mnoP'.title())
print('abcdefg'.replace('cde', 'CDE'))
print('abcdefg'.replace('CDE', 'CDE'))
print('abcdefgabcdefgabcdefg'.replace('cde', 'CDE', 2))
print('  xx  xx  '.strip())
print('\t\r\nxx\t\txx\r\r\r'.strip())
print('xxxxyyyyxxxxxyyyy'.lstrip('x'))
print('xxxxyyyyxxxxxyyyy'.rstrip('y'))
print('xxxxyyyyzxxxxxyyyy'.strip('xy'))
print('1234'.zfill(8))
print('hoge'.ljust(8))
print('hoge'.rjust(8))
print('hoge'.center(8))
```

```python
# その他の文字列メソッド
print('hoge'.find('o'))
print('hogehogehoge'.find('o'))
print('hogehogehoge'.find('z'))
print('HOGEHOGHOGE'.find('h'))
print('hogehogehoge'.split('o'))
print('hoge'.join('fuga'))
print('hogehogehoge'.startswith('ho'))
print('hogehogehoge'.endswith('ho'))
print('hogehogehoge'.encode(encoding='cp932'))
```

```python
# 文字列定数
import string

print(string.ascii_lowercase)
print(string.ascii_uppercase)
print(string.ascii_letters)
print(string.digits)
print(string.hexdigits)
print(string.octdigits)
print(string.punctuation)
print(string.whitespace)
print(string.printable)
```
